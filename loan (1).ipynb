```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score

# Load dataset
url = "https://raw.githubusercontent.com/prasertcbs/basic-dataset/refs/heads/master/Loan-Approval-Prediction.csv"
df = pd.read_csv(url)

# Display first few rows
print(df.head())

# Display dataset shape
print("Shape of dataset:", df.shape)

# Display summary statistics
print(df.describe())

# Check for missing values
print("Missing values:\n", df.isnull().sum())

# Drop missing values
df = df.dropna()

# Replace loan status values with numerical representation
df.replace({"Loan_Status": {"N": 0, "Y": 1}}, inplace=True)

# Verify changes
print(df.head())
```

```python
df["Dependents"].value_counts()
```

```python
df = df.replace(to_replace='3+', value=4)
```

```python
df["Dependents"].value_counts()
```

```python
sns.countplot(x='Education', hue='Loan_Status', data=df)
```

```python
sns.countplot(x='Married', hue='Loan_Status', data=df)
```

```python
df.replace(
    {
        "Married": {"No": 0, "Yes": 1},
        "Gender": {"Male": 0, "Female": 1},
        "Self_Employed": {"No": 0, "Yes": 1},
        "Property_Area": {"Rural": 0, "Semiurban": 1, "Urban": 2},
        "Education": {"Not Graduate": 0, "Graduate": 1}
    },
    inplace=True
)
```

```python
df.head()
```

```python
X= df.drop(columns=["Loan_ID", "Loan_Status"])
y= df ["Loan_Status"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
from sklearn import svm

# Define the classifier
classifier = svm.SVC(kernel='linear')  

# Train the classifier using training data
classifier.fit(X_train, y_train)  
```

```python
y_pred = classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
def predict_loan_approval(input_data):
    input_array = np.array(input_data).reshape(1, -1)
    prediction = classifier.predict(input_array)
    return "Loan Approved" if prediction[0] == 1 else "Loan Not Approved"

# Accept user input
print("Enter the following details:")
gender = int(input("Gender (Male: 0, Female: 1): "))
married = int(input("Married (No: 0, Yes: 1): "))
education = int(input("Education (Not Graduate: 0, Graduate: 1): "))
self_employed = int(input("Self Employed (No: 0, Yes: 1): "))
property_area = int(input("Property Area (Rural: 0, Semiurban: 1, Urban: 2): "))
applicant_income = float(input("Applicant Income: "))
coapplicant_income = float(input("Coapplicant Income: "))
loan_amount = float(input("Loan Amount: "))
loan_amount_term = float(input("Loan Amount Term: "))
credit_history = int(input("Credit History (0 or 1): "))
dependents = int(input("Number of Dependents (Enter 4 for 3+): "))

user_input = [gender, married, education, self_employed, property_area, 
              applicant_income, coapplicant_income, loan_amount, 
              loan_amount_term, credit_history, dependents]

print("Prediction:", predict_loan_approval(user_input))
```
